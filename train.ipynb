```python
# Новый датафрейм с кол-вом строк, как в старом датафрейме. Join по PassengerId.
# Создадим новую колонку evenId,четный или не четный Id. Apply применим к каждому Id в Series. Lambda функция проверит на четность.
mdf = pd.DataFrame(index=df.index)
mdf['PassengerId'] = df['PassengerId']
mdf['evenId'] = mdf['PassengerId'].apply(lambda x: x % 2 == 0) 
```
